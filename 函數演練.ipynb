```python
import numpy as np

def P(x):
    return (1/ (1+ np.exp(-x) ))
  
X = np.arange(-6, 6.1, 0.1)
y = P(X) >= 0.5 
```

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt

def p(t):
    return (np.where(1/(1+np.exp(-t))>=0.5, 1, 0))
    #精華上述使用np.where 批次判斷
    
# 完全保留
T= np.arange(-6, 6.1, 0.1)
y= p(T)
plt.plot(T, y)

plt.grid()
plt.show()
```
