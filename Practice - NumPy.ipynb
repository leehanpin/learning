```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(0, 5, 50)
y = np.linspace(0, 5, 50)[:, np.newaxis]

z = np.sin(x)** 10+ np.cos(10+ y* x)* np.cos(x)
# origin='lower':由小到大，extent=[0, 5, 0, 5]:由0到5
plt.imshow(z,origin='lower', extent=[0, 5, 0, 5],  cmap='viridis')
plt.colorbar();
```

```python
import seaborn; seaborn.set()

mean = [0, 0]
cov = [[1,2],
      [2,5]]
X = np.random.multivariate_normal(mean, cov, 100)#常態分布

plt.scatter(X[:, 0], X[:, 1])
```

```python
indices = np.random.choice(X.shape[0], 20, replace=False)#以X總組數為底，取20個不重複位置
selection = X[indices]
selection.shape
```

```python
plt.scatter(X[:, 0], X[:, 1], alpha=0.5)
plt.scatter(selection[:, 0], selection[:, 1],facecolor='r', s=200 ,alpha=0.3)
```

```python
np.random.seed(42)
x = np.random.randn(100)

bins = np.linspace(-5, 5, 20)
counts = np.zeros_like(bins)#設定與bins相同type和shape的值

#對每一個X找到適當的箱子
i = np.searchsorted(bins, x)

#對這些箱子的每一個家1
np.add.at(counts, i, 1)

plt.plot(bins, counts, linestyle='steps')
```

```python
rand = np.random.RandomState(42)
X = rand.rand(10, 2)
print(X, '\n', '-'*10, '\n', X[:, np.newaxis, :], '\n', '-'*10, '\n', X[np.newaxis, :, :])
```

```python
%matplotlib inline
import matplotlib.pyplot as plt
import seaborn; seaborn.set()

plt.scatter(X[:, 0], X[:, -1], s=100)
```

```python
#計算點與點之間的距離
dist_sq = np.sum((X[:, np.newaxis, :]- X[np.newaxis, :, :])** 2, axis=-1)
dist_sq
```

```python
#點與點之間的距離，分解式
#對每一對點，計算他們的標準差
differences = X[:, np.newaxis, :]- X[np.newaxis, :, :]
print(differences, '\n', '-'*10, '\n', differences.shape)
```
